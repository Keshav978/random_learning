```python
import pandas as pd
import re
import collections
```

```python
df=open("paragraph.txt","r")
```

```python
paragraph_lines = df.readlines()
```

```python
paragraph_lines = [line.strip() for line in paragraph_lines]
```

```python
import string
string.punctuation
```

```python
paragraph_lines=[re.sub( r"[^\w\s]", "", line) for line in paragraph_lines]
```

```python
lower_case_paragraph = [string.lower() for string in paragraph_lines]
```

```python
lower_case_paragraph
```

```python
word_counts = collections.defaultdict(int)
for words in lower_case_paragraph :
  for word in words.split() :
    word_counts[word] += 1
word_counts
```

```python
filesFrame = pd.DataFrame(word_counts.items(), columns=['words','frequency'])
print(filesFrame)
filesFrame[filesFrame["words"] == "a"]
```

```python
filesFrame=filesFrame.sort_values(by="frequency",ascending=False)
frequency_top_five_words=filesFrame.head()
frequency_top_five_words
```
