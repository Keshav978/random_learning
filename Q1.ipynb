```python
import pandas as pd
```

```python
df =pd.read_excel('Data.xlsx')
df
```

```python
bright_skill = df.iloc[:, 3:11].sum(axis = 1)
dark_skill=df.iloc[:,11:15].sum(axis=1)
df["Bright_traits"]= bright_skill
df["Dark_trait"]=dark_skill
df["Bright_traits_avg"]= bright_skill/7
df["Dark_trait_avg"]=dark_skill/4
Bright_traits_cat = []
Dark_trait_cat = []
for ind in df.index:
  if df['Bright_traits_avg'][ind] < 2:
    Bright_traits_cat.append('A')
  elif df['Bright_traits_avg'][ind] < 4:
    Bright_traits_cat.append('B')
  elif df['Bright_traits_avg'][ind] < 7:
    Bright_traits_cat.append('C')
  elif df['Bright_traits_avg'][ind] <= 10:
    Bright_traits_cat.append('D')

  if df['Dark_trait_avg'][ind] < 2:
    Dark_trait_cat.append('A')
  elif df['Dark_trait_avg'][ind] < 4:
    Dark_trait_cat.append('B')
  elif df['Dark_trait_avg'][ind] < 7:
    Dark_trait_cat.append('C')
  elif df['Dark_trait_avg'][ind] <= 10:
    Dark_trait_cat.append('D')

df["Dark_trait_cat"] = Dark_trait_cat
df["Bright_traits_cat"] = Bright_traits_cat
df
```
